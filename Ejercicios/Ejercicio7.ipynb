Si cliente es VIP, 15% en la compra

```python
class cliente:

    def __init__(self,nombre,vip=False):
        self.nombre = nombre
        self.vip = vip
        self.cesta = []

    def pagar(self,precio):
        print("El cliente {} ha pagado {}".format(self.nombre,precio))
        

class cajera:

    def __init__(self,nombre):
        self.nombre = nombre

    def atender_cliente(self,cliente): 
        importe = 0
        for producto in cliente.cesta:
            importe += producto.precio
        
        if cliente.vip:
            print("Descuento a usuario vip del 15%")
            importe -= (importe*0.15)
        
        return importe



        


class tienda:

    def __init__(self,cajeros,productos):
        self.clientes = []
        self.cajeros = cajeros
        self.productos = productos

    def cobrar_cliente(self):
        hay_cliente = self.hay_clientes()
        while (hay_cliente):
            for cajera in self.cajeros:
                if self.hay_clientes():
                    cliente = self.clientes.pop(0)
                    print("Esta atendiendo al cliente {} por la cajera {}".format(cajera.nombre,cliente.nombre))
                    importe = cajera.atender_cliente(cliente)
                    cliente.pagar(importe)
                    hay_cliente = self.hay_clientes()



        
    def hay_clientes(self):
        return len(self.clientes) >0
    
    
        

class television:

    def __init__(self,id,tipo,marca,pulgadas,precio):
        self.id = id
        self.tipo = tipo
        self.marca = marca
        self.pulgadas = pulgadas
        self.precio = precio


class ordenador:

    def __init__(self,id,tipo,marca,precio):
        self.id = id
        self.tipo = tipo
        self.marca = marca
        self.precio = precio

class monitor:

    def __init__(self,id,tipo,marca,pulgadas,precio):
        self.id = id
        self.tipo = tipo
        self.marca = marca
        self.pulgadas = pulgadas
        self.precio = precio




monitor1 = monitor("1A","monitor","Samsung",22,180)
ordenador1 = ordenador("2B","ordenador","HP",290)
television1 = television("3C","television","LG",32,210)
productos = [monitor1,ordenador1,television1]
cliente1 = cliente("alvaro",True)
cliente2 = cliente("pepe")
cliente3 = cliente("raul")
cajero1 = cajera("Maria")
cajero2 = cajera("Luis")
cajeros = [cajero1,cajero2]

tienda1 = tienda(cajeros,productos)
tienda1.clientes.append(cliente1)
tienda1.clientes.append(cliente2)
tienda1.clientes.append(cliente3)

cliente1.cesta.append(monitor1)
cliente2.cesta.append(ordenador1)
cliente3.cesta.append(television1)

tienda1.cobrar_cliente()

        

        
        
```
