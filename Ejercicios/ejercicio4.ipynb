```python
opcion = 0
while opcion != 4:
    print("##################")
    print("1. Pintar cuadrado")
    print("2. Validar email")
    print("3. Insertar alumnos")
    print("4. Salir")
    print("##################")
    opcion = int(input("ELige una opcion: "))
    if opcion == 1:
        caracter = "*"
        altura = int(input("\nIndica tamaño altura"))
        base = int(input("\nIndica tamaño base"))
        for linea in range(altura):
            if(linea==0 or linea == altura-1):
                print(base*caracter)
            else:
                print(caracter, end = "")
                cadena = " " * (base-2)
                print(cadena, end="")
                print(caracter)
                
    elif opcion == 2:
        #lista_emails = ["pep egmail.com","prueba2@gmail.com","javier .m"]
        lista_emails = ["asdasdas.es","asdas asdsdsa@sda.e","asdasd@asdas@asdsad.es", "asdad@a.es", "a sgsdfsd@a.esfa.sfasdf", "asdas.dasd@as.dsfd.as.es "]
        for email in lista_emails:
            email = email.strip()
            print("Validando email ", email)
            
            errores = ""
            if email.count(" ")>0:
                errores += "No puede haber espacios en blanco. "
            if not email.count("@")==1:
                errores += "Debe de haber una @. "
            if not email.split("@")[-1].count(".")>=1:
                errores += "Despues de la @ tiene que haber un punto. "
            if not email[ email.find("@")+1 :  ].find(".")>1:
                errores += "La distancia entre la @ y el primer punto despues de la @ debe ser mayor de 2. "
            if len(email.split(".")[-1])<2 or len(email.split(".")[-1])>5:
                errores += "El subdominio del email debe estar entre 2 y 5. "
            
            if (errores==""):
                print("El email "+ email +" es correcto.")
            else:
                print("El email "+ email +" no es correcto " + errores)

        
    elif opcion == 3:
        # 1. Insertar aulmno --> En que aula quieres meter al alumno (1,2,3) --> DNI, Nombre, Apellidos 
        # 2. Mostrar alumnos colegio --> Todos los alumnos del colegio por aula
        # 3. Mostrar alumnos aula --> Pregunta de que aula quiere mostrar los alumnos (1,2,3)
        # 4. Buscar alumno --> Pregunta por el DNI --> Muestra en el aula que esta el alumno con ese DNI, si existe
        # 5. Salir

        # Tenemos un colegio en el que hay clases a las que estan asignadas unos alumos con DNI, NOMBRE Y APELLIDOS

        colegio = {}
        aulas = [1,2,3]
        alumnos = []

        def menu():
            opcion = 0
            while opcion != 5:
                print("1. Insertar aulmno")
                print("2. Mostrar alumnos colegio")
                print("3. Mostrar alumnos aula")
                print("4. Busvar alumnos")
                print("5. Salir")
                opcion = int(input("Elige una opcion: "))
                if opcion == 1:
                    insertar_alumno()
                if opcion == 2:
                    print(mostrar_alumnos())
                if opcion == 3:
                    print(mostrar_alumnos_aula(int(input("Elige la aula para mostrar alumnos"))))
                if opcion == 4:
                    print(buscar_alumno(str(input("Escribe Dni para buscar alumno"))))
                if opcion == 5:
                    print("Adios!!!")

        def insertar_alumno():
            dni = str(input("Inserta Dni alumno: "))
            nombre = str(input("Inserta nombre alumno: "))
            apellidos = str(input("Inserta apellidos alumno: "))
            aula = int(input("Inserta aula alumno: ")) 
            alumno = {'Dni': dni, 'Nombre': nombre,'Apellidos': apellidos, "Aula": aula-1 }
            alumnos.append(alumno)
            colegio['Alumno','Aulas'] = alumnos,aulas


        def mostrar_alumnos():    
            aula1 = ""
            aula2 = ""
            aula3 = ""
            for alumno in alumnos:
                if alumno['Aula'] == 0:
                    aula1 += str(alumno)+", " 
                if alumno['Aula'] == 1:
                    aula2 += str(alumno)+", " 
                if alumno['Aula'] == 2:
                    aula3 += str(alumno)+", " 

            return str("Alumnos aula 1: "+aula1+"\n"+"Alumnos aula 2: "+aula2+"\n"+"Alumnos aula 3: "+aula3+"\n")    


        def mostrar_alumnos_aula(aula):
            alumnos_aula = ""
            for alumno in alumnos:
                if alumno['Aula'] == aula-1:
                    alumnos_aula += str(alumno)+", "
                if aula >= 4:
                    return "Aula no existente"
            if alumnos_aula != "":
                return alumnos_aula
            else:
                return "Aula vacia"


        def buscar_alumno(dni):
            alumno_dni = ""
            for alumno in alumnos:
                if alumno['Dni'] == dni:
                    alumno_dni = str(alumno)

            if alumno_dni != "":
                return alumno_dni
            else:
                return "Alumno no encontrado"
        menu()

    elif opcion == 4:
        print("Adios")
    else:
        print("Opcion incorrecta")

    
```
